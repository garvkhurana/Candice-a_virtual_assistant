```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"C:\Users\garvk\OneDrive - Bhagwan Parshuram Institute of Technology\Desktop\advance_projects\candice\runs\detect\yolo11s_finetuned4\results.csv")
df.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 7))
sns.lineplot(ax=axes[0], x=df.index, y="train/box_loss", data=df, label="Box Loss", linewidth=2)
sns.lineplot(ax=axes[0], x=df.index, y="train/cls_loss", data=df, label="Cls Loss", linewidth=2)
sns.lineplot(ax=axes[0], x=df.index, y="train/dfl_loss", data=df, label="DFL Loss", linewidth=2)

axes[0].set_title("YOLO Training Losses", fontsize=14, weight="bold")
axes[0].set_xlabel("Epochs", fontsize=12)
axes[0].set_ylabel("Loss", fontsize=12)
axes[0].legend()
axes[0].grid(True)



sns.lineplot(ax=axes[1], x=df.index, y="val/box_loss", data=df, label="Val Box Loss", linewidth=2)
sns.lineplot(ax=axes[1], x=df.index, y="val/cls_loss", data=df, label="Val Cls Loss", linewidth=2)
sns.lineplot(ax=axes[1], x=df.index, y="val/dfl_loss", data=df, label="Val DFL Loss", linewidth=2)

axes[1].set_title("YOLO Validation Losses", fontsize=14, weight="bold")
axes[1].set_xlabel("Epochs", fontsize=12)
axes[1].set_ylabel("Loss", fontsize=12)
axes[1].legend()
axes[1].grid(True)


plt.suptitle("YOLO Training vs Validation Loss Trends", fontsize=16, weight="bold")
plt.tight_layout()
plt.show()
```

```python
fig1, axes1 = plt.subplots(1, 2, figsize=(18, 7))
sns.lineplot(ax=axes1[0], x=df.index, y="metrics/precision(B)", data=df, label="precision", linewidth=2)
sns.lineplot(ax=axes1[0], x=df.index, y="metrics/recall(B)", data=df, label="recall", linewidth=2)

axes1[0].set_title("precision_and_recall_graph", fontsize=14, weight="bold")
axes1[0].set_xlabel("Epochs", fontsize=12)
axes1[0].set_ylabel("precision_and_recall", fontsize=12)
axes1[0].legend()
axes1[0].grid(True)



sns.lineplot(ax=axes1[1], x=df.index, y="metrics/mAP50(B)", data=df, label="map50", linewidth=2)
sns.lineplot(ax=axes1[1], x=df.index, y="metrics/mAP50-95(B)", data=df, label="mAP50-95", linewidth=2)

axes1[1].set_title("map_gain_graph", fontsize=14, weight="bold")
axes1[1].set_xlabel("Epochs", fontsize=12)
axes1[1].set_ylabel("MAP_gain", fontsize=12)
axes1[1].legend()
axes1[1].grid(True)


plt.suptitle("YOLO Training and precision_and_recall_graph and map_gain_graph", fontsize=16, weight="bold")
plt.tight_layout()
plt.show()
```
